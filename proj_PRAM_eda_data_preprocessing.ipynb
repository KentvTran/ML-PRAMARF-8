```python
#Information Gain
import pandas as pd
from sklearn.metrics import mutual_info_score
from sklearn.preprocessing import LabelEncoder

csvfile_path = 'C:/Users/kentv/OneDrive/Desktop/ML/ML Proj - PRAM/Data/raw/cleanPhase8ARFv1forEDA.csv'
data = pd.read_csv(csvfile_path)

# Define the features (X) and the target variable (y)
X = data.drop(columns=["INF_LIVE"]) 
y = data["INF_LIVE"]

# Apply label encoding to Nominal Variables
columns_to_encode = ["STATE", "MAT_RACE_PU", "PAT_RACE_PU", "PAY"]
X_encoded = X.copy()

le = LabelEncoder()
for col in columns_to_encode:
    X_encoded[col] = le.fit_transform(X[col].astype(str))

# Calculate Information Gain (IG) for each feature
def information_gain(feature, target):
    total_entropy = mutual_info_score(target, target)
    conditional_entropy = total_entropy - mutual_info_score(feature, target)
    return total_entropy - conditional_entropy

ig_results = pd.DataFrame({'Feature': X.columns})
ig_results['Information Gain'] = [information_gain(X_encoded[col], y) for col in X_encoded.columns]

print("\nInformation Gain (IG) Scores:")
print(ig_results.sort_values(by="Information Gain", ascending=False))
```

```python
#Principal Component Analysis (PCA)
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

csvfile_path = 'C:/Users/kentv/OneDrive/Desktop/ML/ML Proj - PRAM/Data/raw/cleanPhase8ARFv1forEDA.csv'
data = pd.read_csv(csvfile_path)

selected_features = [
    "HISPANIC", "MARRIED", "MAT_WIC", "PAT_ACK", "BPG_DIAB8", "BPG_HBP8", "BPG_DEPRS8",
    "HTH_ASMA", "HTH_THYR", "HTH_PCOS", "HTH_ANX", "TLK_WT_RAW", "ASK_MEDS_RAW",
    "TLK_CIGS_RAW", "TLK_ETOH_RAW", "ASK_ABUS_RAW", "ASK_DPRS_RAW", "ASK_DRUG_RAW",
    "ASK_HIVT_RAW", "ASK_BF_RAW", "ASK_PPBC_RAW", "PG_GDB8", "MORB_BP8", "MH_PGDX8",
    "PAB6HUS", "PAB_XHUS", "PAB_FAM", "PAB_OTH", "PAD6HUS", "PAD_XHUS", "PAD_FAM", "PAD_OTH", 
    "MAT_DEG", "PAT_DEG", "PAY", "STATE", "MAT_RACE_PU", "PAT_RACE_PU", "INCOME8", "MOM_LBKG", "ILLB_MO"
]
target_variable = 'INF_LIVE'

X = data[selected_features]

# Target encoding for Nominal variables and STATE
for col in ["PAY", "MAT_RACE_PU", "PAT_RACE_PU", "STATE"]:
    target_means = data.groupby(col)[target_variable].mean()
    X.loc[:, col] = X[col].map(target_means)

# Drop rows with NaN values introduced during target encoding (unmapped categories)
X = X.dropna(subset=["STATE", "PAY", "MAT_RACE_PU", "PAT_RACE_PU"])

# Standardize the features before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to the scaled data
pca = PCA()
pca.fit(X_scaled)

# Create a DataFrame to display the explained variance ratio
explained_variance = pd.DataFrame(
    pca.explained_variance_ratio_,
    columns=["Explained Variance"],
    index=[f"PC{i+1}" for i in range(len(selected_features))],
)

# Create a DataFrame for the loadings
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f"PC{i+1}" for i in range(len(selected_features))],
    index=selected_features,
)

print("\nExplained Variance Ratio for Each Principal Component:")
print(explained_variance)

print("\nPCA Loadings (Contribution of Each Feature to Principal Components):")
print(loadings)

# Function to get the top contributing variables for each principal component
def get_top_contributors(loadings, n_top=10):
    top_contributors = {}
    
    # For each principal component
    for pc in loadings.columns[:5]:  # Limiting to first 5 PCs (PC1 to PC5)
        # Sort the features by the absolute value of their loadings in descending order
        sorted_vars = loadings[pc].abs().sort_values(ascending=False)
        # Get the top n contributing variables
        top_contributors[pc] = sorted_vars.head(n_top)
    
    return top_contributors

top_contributors = get_top_contributors(loadings)

# Display the top contributing variables for each PC
print("\nTop Contributing Variables for Each Principal Component:")
for pc, vars in top_contributors.items():
    print(f"{pc}:")
    print(vars)
    print()
```

```python
#Correlation Between Variable of Interest and Target
import pandas as pd
from scipy.stats import pointbiserialr, chi2_contingency, spearmanr
import numpy as np

csvfile_path = 'C:/Users/kentv/OneDrive/Desktop/ML/ML Proj - PRAM/Data/raw/cleanPhase8ARFv1forEDA.csv'
data = pd.read_csv(csvfile_path)

binary_categorical_vars = [
    "HISPANIC", "MARRIED", "MAT_WIC", "PAT_ACK", "BPG_DIAB8", "BPG_HBP8", "BPG_DEPRS8",
    "HTH_ASMA", "HTH_THYR", "HTH_PCOS", "HTH_ANX", "TLK_WT_RAW", "ASK_MEDS_RAW",
    "TLK_CIGS_RAW", "TLK_ETOH_RAW", "ASK_ABUS_RAW", "ASK_DPRS_RAW", "ASK_DRUG_RAW",
    "ASK_HIVT_RAW", "ASK_BF_RAW", "ASK_PPBC_RAW", "PG_GDB8", "MORB_BP8", "MH_PGDX8",
    "PAB6HUS", "PAB_XHUS", "PAB_FAM", "PAB_OTH", "PAD6HUS", "PAD_XHUS", "PAD_FAM", "PAD_OTH"
]

ordinal_categorical_vars = ["INCOME8","MAT_DEG", "PAT_DEG", "PAY"]
regular_categorical_vars = ["STATE", "MAT_RACE_PU", "PAT_RACE_PU"]
numerical_vars = ["MOM_LBKG", "ILLB_MO"]

target_var = "INF_LIVE"

# Ensure target variable is numeric for calculations
data[target_var] = data[target_var].astype(int)

# Point Biserial Correlation for numerical variables
print("Point Biserial Correlation (Numerical vs Target Variable):")
for var in numerical_vars:
    if var != target_var:  
        corr, p_value = pointbiserialr(data[target_var], data[var])
        print(f"{var} - Correlation: {corr:.3f}, p-value: {p_value:.3f}")

# Chi-square test for binary categorical
print("\nChi-Square Test (Binary Categorical vs Target Variable):")
for var in binary_categorical_vars:
    contingency_table = pd.crosstab(data[target_var], data[var])
    chi2, p, dof, ex = chi2_contingency(contingency_table)
    print(f"{var} - Chi2: {chi2:.3f}, p-value: {p:.3f}")

# Chi-square test for regular categorical variables
print("\nChi-Square Test (Regular Categorical vs Target Variable):")
for var in regular_categorical_vars:
    contingency_table = pd.crosstab(data[target_var], data[var])
    chi2, p, dof, ex = chi2_contingency(contingency_table)
    print(f"{var} - Chi2: {chi2:.3f}, p-value: {p:.3f}")

# Spearman Correlation for ordinal categorical variables
print("\nSpearman Correlation (Ordinal Categorical vs Target Variable):")
for var in ordinal_categorical_vars:
    corr, p_value = spearmanr(data[target_var], data[var])
    print(f"{var} - Spearman Correlation: {corr:.3f}, p-value: {p_value:.3f}")
```

```python
#Correlation Between Variables (Removed Low Info Gain)
import pandas as pd
import itertools
from scipy.stats import chi2_contingency

# Load the dataset
csvfile_path = 'C:/Users/kentv/OneDrive/Desktop/ML/ML Proj - PRAM/Data/raw/cleanPhase8ARFv1forEDA.csv'
data = pd.read_csv(csvfile_path)

# Define variable groups
binary_categorical_vars = [
    "ASK_PPBC_RAW", "ASK_BF_RAW",  
    "HTH_PCOS", "HTH_ANX", "MH_PGDX8", "HTH_THYR",  
    "TLK_CIGS_RAW", "TLK_ETOH_RAW",   
    "PAT_ACK", "PAD_FAM", "MAT_WIC"
]

ordinal_categorical_vars = ["INCOME8", "MAT_DEG", "PAT_DEG"]
regular_categorical_vars = ["STATE", "MAT_RACE_PU", "PAT_RACE_PU", "PAY"]
numerical_vars = ["MOM_LBKG", "ILLB_MO"]

# Initialize correlation matrices as DataFrames
pearson_corr_matrix = pd.DataFrame(index=numerical_vars, columns=numerical_vars)
spearman_corr_matrix = pd.DataFrame(index=ordinal_categorical_vars, columns=ordinal_categorical_vars)
chi2_corr_matrix = pd.DataFrame(index=binary_categorical_vars + regular_categorical_vars, columns=binary_categorical_vars + regular_categorical_vars)

# Pearson Correlation for Numerical vs. Numerical
for var1, var2 in itertools.combinations(numerical_vars, 2):
    corr = data[[var1, var2]].corr().iloc[0, 1]  # Pearson correlation
    pearson_corr_matrix.loc[var1, var2] = corr
    pearson_corr_matrix.loc[var2, var1] = corr

# Spearman Correlation for Ordinal Categorical vs. Ordinal Categorical
for var1, var2 in itertools.combinations(ordinal_categorical_vars, 2):
    corr = data[[var1, var2]].corr(method='spearman').iloc[0, 1]  # Spearman correlation
    spearman_corr_matrix.loc[var1, var2] = corr
    spearman_corr_matrix.loc[var2, var1] = corr

# Chi-Square Approximation for Binary and Regular Categorical Variables
for var1, var2 in itertools.combinations(binary_categorical_vars + regular_categorical_vars, 2):
    contingency_table = pd.crosstab(data[var1], data[var2])
    chi2 = chi2_contingency(contingency_table)[0]  # Chi-square value (proxy for correlation strength)
    chi2_corr_matrix.loc[var1, var2] = chi2
    chi2_corr_matrix.loc[var2, var1] = chi2

# Fill diagonal with 1 (perfect self-correlation)
pearson_corr_matrix.fillna(1, inplace=True)
spearman_corr_matrix.fillna(1, inplace=True)
chi2_corr_matrix.fillna(1, inplace=True)

# Output results
print("Pearson Correlation Matrix for Numerical Variables:")
print(pearson_corr_matrix)

print("\nSpearman Correlation Matrix for Ordinal Variables:")
print(spearman_corr_matrix)

print("\nChi-Square Approximation Matrix for Binary and Regular Categorical Variables:")
print(chi2_corr_matrix)
```
